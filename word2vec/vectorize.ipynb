```python
from gensim.models import Word2Vec
import gensim
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
import nltk

nltk.download('punkt_tab')

warnings.filterwarnings(action='ignore')
```

```python
import zipfile

cleaned_text = None
with zipfile.ZipFile("./Gutenburg.zip", 'r') as zip_ref:
    file_name = zip_ref.namelist()[0]
    with zip_ref.open(file_name) as file:
        content = file.read().decode('utf-8', errors='ignore')
        cleaned_text = content.replace("\n", " ")
        print("File loaded")
```

```python
import textwrap
first_200_words = " ".join(cleaned_text.split()[:200])

print(textwrap.fill(first_200_words, width=80))
```

```python
data = []

for i in sent_tokenize(cleaned_text):
    print(i)
    temp = []

    for j in word_tokenize(i):
        temp.append(j.lower())

    data.append(temp)
```

```python
model1 = gensim.models.Word2Vec(data, min_count=1, vector_size=100, window=5)
model2 = gensim.models.Word2Vec(data, min_count=1, vector_size=100, window=5, sg=1)
```

```python
print("Cosine similarity between 'alice' and 'wonderland' - CBOW : ", model1.wv.similarity('alice', 'wonderland'))
print("Cosine similarity between 'alice' and 'chapter' - CBOW : ", model1.wv.similarity('alice', 'chapter'))
print("Cosine similarity between 'alice' and 'hate' - CBOW : ", model1.wv.similarity('alice', 'hate'))
```

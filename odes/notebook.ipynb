```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(
    np.linspace(-3, 3, 25),
    np.linspace(-3, 3, 25)
)

dy_dx = x/y

# direction vectors (dx, dy)
dx = np.ones_like(dy_dx)
dy = dy_dx

# normalize so arrows show direction only
norm = np.sqrt(dx**2 + dy**2)
dx /= norm
dy /= norm

plt.quiver(x, y, dx, dy)
plt.xlabel("x")
plt.ylabel("y")
# plt.axis("equal")
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(
    np.linspace(-3, 3, 25),
    np.linspace(-3, 3, 25)
)

dy_dx = 2*x

# direction vectors (dx, dy)
dx = np.ones_like(dy_dx)
dy = dy_dx

# normalize so arrows show direction only
norm = np.sqrt(dx**2 + dy**2)
dx /= norm
dy /= norm

plt.quiver(x, y, dx, dy)
plt.xlabel("x")
plt.ylabel("y")
# plt.axis("equal")
plt.show()


```

```python
import matplotlib.pyplot as plt


def show_img(path, return_img=False):
    img = plt.imread(path)
    plt.imshow(img)
    plt.axis("off")
    plt.show()

    if return_img: return img


show_img("./all-of-the-lights.png")
```

```python
img = show_img("./lincoln.png", return_img=True)
```

```python
from scipy.ndimage import gaussian_filter

new_img = gaussian_filter(img, sigma=1)
plt.imshow(new_img)
plt.axis("off")
plt.show()
```

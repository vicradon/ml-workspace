```
import torch
from torch import nn

class XORPytorchNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(2, 4)  # Hidden layer with 4 neurons
        self.fc2 = nn.Linear(4, 1)  # Output layer
        
    def forward(self, x):
        x = torch.relu(self.fc1(x))  # Non-linear activation
        x = torch.sigmoid(self.fc2(x))  # Sigmoid for binary output
        return x

# XOR training data
X = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)
y = torch.tensor([[0], [1], [1], [0]], dtype=torch.float32)

# Initialize model and loss
model = XORPytorchNetwork()
criterion = nn.BCELoss()  # Binary Cross Entropy Loss

# Hyperparameters
learning_rate = 0.1
epochs = 5000

# Training loop with manual gradient descent
for epoch in range(epochs):
    # Forward pass
    outputs = model(X)
    loss = criterion(outputs, y)
    
    # Backward pass
    loss.backward()
    
    # Manual gradient descent update
    with torch.no_grad():
        for param in model.parameters():
            param -= learning_rate * param.grad  # Update: θ = θ - α * ∇θ
            param.grad.zero_()  # Clear gradients for next iteration
    
    # Print progress every 1000 epochs
    if (epoch + 1) % 1000 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# Test the trained model
print("\nTest Results:")
with torch.no_grad():
    predictions = model(X)
    for i in range(len(X)):
        print(f'Input: {X[i].numpy()} -> Prediction: {predictions[i].item():.4f} (Target: {y[i].item()})')
```
